```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 拆线图

```python
x=np.linspace(0, 2*np.pi, 100)

plt.plot(x, np.sin(x), color='blue')
plt.plot(x, np.sin(x-np.pi/3), color='r')
plt.plot(x, np.sin(x-2*np.pi/3), color='#a52a2a')
```

```python
plt.plot(x, 2*x, linestyle='dashdot')
plt.plot(x, -2*x+2, linestyle=':')
```

```python
plt.plot(range(10), linestyle='--', color='r',
         marker='o', markersize=14, markerfacecolor='b')
plt.grid(True)
```

```python
x=np.linspace(-np.pi, np.pi, 9)
plt.plot(x, np.sin(x), '-Dr',
         markersize=16, markerfacecolor='b', markevery=[2, 4, 6, 7])
plt.plot(x, np.cos(x), 'hr',
         markersize=14, markerfacecolor='m')
plt.grid(True)
```

```python
a=np.arange(0,3, .02)
b=np.arange(0,3, .02)
c=np.exp(a)
d=c[::-1]

plt.plot(a, c, 'k--', label='Model')
plt.plot(a, d, 'r:', label='Data')
plt.plot(a, c+d, 'b-', label='Total')

plt.legend(loc='upper center')
```

```python
# plot第一个对象为线对象
line1, =plt.plot(a, c, 'k--', label='Model')
line2=plt.plot(a, d, 'g:', label='Data')[0]
line3, =plt.plot(a, c+d, 'b-', label='Total')

plt.legend((line1, line2), loc=0)
```

## 散点图

```python
rng=np.random.RandomState(0)
x=rng.randn(100)
y=rng.randn(100)

colors=rng.rand(100)
sizes=1000*rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3)
plt.colorbar()
```

## 柱状图

```python
data=[2, 10, 4, 8, 6]
position=range(1,6)
plt.bar(x=position, height=data)
```

```python
data=[2, 10, 4, 8, 6]
position=range(1,6)
labels = ['Pejing', 'Soochow', 'Hongkong', 'Tsingtao', 'Canton']
plt.bar(position, data, width=0.4,
        color='gb', edgecolor='r', linestyle='--',
        linewidth=3, hatch='x', tick_label=labels)
```

```python
position=range(1, 7)
a=np.random.random(6)
b=np.random.random(6)
plt.bar(position, a, label='a', color='k')
# 在a的数据上叠加
plt.bar(position, b, bottom=a, label='b', color='r')
plt.legend()
```

## 直方图

```python
datas=np.random.randn(100000)
hs=plt.hist(datas)
```

```python
from scipy.stats import norm
fig=plt.figure()
mu=100  # 平均值
sigma=15  # 标准差
x=mu+sigma*np.random.randn(10000)
num_bins=50

# 绘制频率直方图
n, bins, patches = plt.hist(x, num_bins, normed=True, facecolor='blue', alpha=0.6)

# 绘制正态分布曲线
y=norm.pdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')
```
