### 基于热度的推荐算法

```python
ratings_sum=pd.DataFrame(ratings.groupby('isbn').rating.sum())
ratings_sum
```

```python
reading_count=pd.DataFrame(ratings.groupby('isbn').user_id.count().sort_values(ascending=False))
reading_count.head(20)
```

```python
avg_ratings=pd.DataFrame(ratings.groupby('isbn').rating.mean().sort_values(ascending=False))
avg_ratings.head(60)
```

```python
top10=ratings_sum.sort_values('rating', ascending=False).head(10)
top10
```

```python
top10.merge(books, right_on='isbn', left_index=True )
```

### 基于用户的协同过滤算法

```python
ratings.user_id.value_counts().describe()
```

```python
ratings.user_id.value_counts().sort_values(ascending=False).head(50)
```

```python
ratings.isbn.value_counts().describe()
```

```python
ratings.isbn.value_counts().sort_values(ascending=False).head(50)
```

```python
counts=ratings.user_id.value_counts()
ratings=ratings[ratings.user_id.isin(counts[counts>10].index)]
ratings
```

```python
counts=ratings.isbn.value_counts()
ratings=ratings[ratings.isbn.isin(counts[counts>10].index)]
ratings
```

```python
# 通过透视表形成矩阵
ratings_matrix=ratings.pivot(index='user_id', columns='isbn', values='rating')
ratings_matrix
```

```python
ratings_matrix.shape
```

```python
user_id=ratings_matrix.index
isbn=ratings_matrix.columns
```

```python
ratings_matrix.fillna(0, inplace=True)
ratings_matrix=ratings_matrix.astype(np.int32)
```
