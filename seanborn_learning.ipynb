### 

`sns.catplot``kind`

- `kind = 'point'``sns.pointplot`

```python
g = sns.catplot(x='class',
                y='survived',
                hue='sex',
                data=titanic,
                height=6,
                kind='bar',
                palette='muted')
g.despine(left=True)
g.set_ylabels('survival probability')
```
